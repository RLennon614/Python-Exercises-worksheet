You are going to use the different Python concepts seen to build a program that will simulate the behaviour of an ATM. Follow the instructions below:

Print a main menu to ask the user different options for displaying the bank account balance, for depositing money, to withdraw money, and to exit from the ATM. Ideally create a function for each menu option to be executed. Use conditional statements for asking the user the different menu options.

Additional (optional): Ask the user a PIN to access to the ATM menu once the PIN is correct otherwise ask the user to try again.

Prompt the user to enter an option:

If ‘1’ is entered, display the current balance and return to main menu.

If ‘2’ is entered, print a sub-menu with different withdrawal amounts (give also a choice for “other amount’ to let the user enter an amount) :

Check that the requested withdrawal is allowed, print a message to show that the money has been withdrawn, calculate the new balance and return to main menu.

If ‘other_amount’ is selected, then prompt the user for an integer value. Check this number is a multiple of 10 and that the withdrawal is permitted, print a message to show that the money has been withdrawn, calculate the new balance and return to main menu.

If ‘exit’ is selected print a goodbye message and exit (break).

If another value is entered, print an error message and print the menu again.



-Re-use the ATM menu options across the program to ask the user if he/she would like to perform another operation on the bank account.

```python
# ATM Simulation Program

# --- Optional: Set a PIN for access ---
PIN = "1234"

# --- Initial account balance ---
balance = 1000.0  # starting balance

# --- Function to check PIN ---
def check_pin():

    for attempt in range(3):
        entered_pin = input("Enter your PIN: ")
        if entered_pin == PIN:
            print("PIN correct. Access granted.\n")
            return True
        else:
            print(" Incorrect PIN. Try again.\n")
    print(" Too many incorrect attempts. Card blocked. Please contact the Bank.")
    return False

# --- Function to display balance ---
def display_balance(balance):
    print(f"\n Your current balance is: ${balance:.2f}\n")
    return balance

 # --- Function to deposit money ---
def deposit_money(balance):
    amount = float(input("Enter amount to deposit: $"))
    if amount > 0:
        balance += amount
        print(f"${amount:.2f} deposited successfully.")
        print(f"New balance: ${balance:.2f}\n")
    else:
        print(" Invalid amount. Deposit cancelled.\n")
    return balance

# --- Function to withdraw money ---
def withdraw_money(balance):
    print("\nSelect withdrawal amount:")
    print("1. $20")
    print("2. $40")
    print("3. $60")
    print("4. $100")
    print("5. $200")
    print("6. Other amount")
    choice = input("Enter your choice (1–6): ")

    if choice in ["1", "2", "3", "4", "5"]:
        amounts = {"1": 20, "2": 40, "3": 60, "4": 100, "5": 200}
        amount = amounts[choice]
    elif choice == "6":
        amount = int(input("Enter withdrawal amount (multiple of 10): "))
        if amount % 10 != 0:
            print(" Amount must be a multiple of 10.\n")
            return balance
    else:
        print(" Invalid selection.\n")
        return balance

    # Check sufficient balance
    if amount <= balance:
        balance -= amount
        print(f" Please take your cash: ${amount:.2f}")
        print(f"Remaining balance: ${balance:.2f}\n")
    else:
        print(" Insufficient balance.\n")
    return balance

 # --- Main ATM menu ---
def atm_menu():
    balance = 1000.0  # starting balance
    while True:
        print("==== WELCOME TO EXERCISE 3 BANK ATM ====")
        print("1. Check Balance")
        print("2. Withdraw Money")
        print("3. Deposit Money")
        print("4. Exit")
        choice = input("Enter your choice (1–4): ")

        if choice == "1":
            balance = display_balance(balance)
        elif choice == "2":
            balance = withdraw_money(balance)
        elif choice == "3":
            balance = deposit_money(balance)
        elif choice == "4":
            print("\n Thank you for banking with Exercise 3 Bank. Goodbye!")
            break
        else:
            print(" Invalid option. Please try again.\n")

        again = input("Would you like another transaction? (yes/no): ").lower()
        if again != "yes":
            print("\n Thank you for using Exercise 3 Bank. Have a nice day!")
            break
        print("\n---------------------------------\n")

 # --- Run the program ---
if check_pin():
    atm_menu()
```
